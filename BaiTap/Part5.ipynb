Sử dụng Pandas để Phân Tích Dữ Liệu

```python
#Bài tập 1: Tạo một DataFrame từ một từ điển chứa thông tin về sinh viên (tên, tuổi, điểm số). In ra DataFrame và tính trung bình điểm số.

import pandas as pd 

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [20, 21, 19],
    'Score': [85, 92, 78]
}

df = pd.DataFrame(data)

print("DataFrame:")
print(df)

# Tính trung bình điểm
average_score = df['Score'].mean()
print("Trung bình điểm số:", average_score)

```

```python
#Bài tập 2: Tạo một DataFrame mới chỉ chứa những sinh viên có điểm số lớn hơn 80 và in ra DataFrame này.

high_scores = df[df['Score']> 80]
print("Sinh viên có điểm số lớn hơn 80:")
print(high_scores)
```

 Sử dụng NumPy để Phân Tích Dữ Liệu

```python
#Bài tập 3: Tạo một mảng NumPy chứa 100 số ngẫu nhiên từ phân phối chuẩn và tính trung bình, độ lệch chuẩn của mảng đó.

import numpy as np

# tạo mảng ngẫu nhiên

data = np.random.randn(100)

#tính trùng bình và độ lệch chuẩn

mean = np.mean(data)
std_dev = np.std(data)

print("Trung bình:", mean)
print("Độ lệch chuẩn:", std_dev)
```

```python
#Bài tập 4: Viết một hàm nhận vào một mảng NumPy và trả về các phần tử lớn hơn trung bình của mảng đó.

def elements_above_mean(arr):
    mean_value = np.mean(arr)
    return arr[arr > mean_value]

arr = np.array([10,20,30,40,50])
print("Các phần tử lớn hơn trung bình:", elements_above_mean(arr))
```

Vẽ Biểu Đồ với Matplotlib

```python
#Bài tập 5: Sử dụng matplotlib để vẽ biểu đồ đường thể hiện điểm số của sinh viên qua thời gian (giả định rằng có 5 bài kiểm tra).

import matplotlib.pyplot as plt

# Dữ liệu giả định
tests = [1,2,3,4,5]
scores = [78,85,90,92,88]

plt.plot(tests, scores, marker = 'o')
plt.title("Điểm số qua các bài kiểm tra")
plt.xlabel("Bài kiểm tra")
plt.ylabel("Điểm số")
plt.xticks(tests)
plt.grid()
plt.show

```

```python
#Bài tập 6: Vẽ biểu đồ cột để so sánh điểm số của 3 sinh viên.

students = ['Alice', 'Bob', 'Charlie']
scores = [85, 92, 78]

plt.bar(students, scores, color=['blue','orange', 'green'])
plt.title("So sánh điểm số của sinh viên")
plt.xlabel("Sinh viên")
plt.ylabel("Điểm số")
plt.show()
```

```python
#Bài tập 7: Vẽ biểu đồ phân tán để thể hiện mối quan hệ giữa hai biến (ví dụ: điểm số và tuổi).

ages = [20, 21, 19, 22, 20]
scores = [85, 92, 78, 88, 90]

plt.scatter(ages, scores, color='red')
plt.title("Mối quan hệ giữa điểm số và tuổi")
plt.xlabel("Tuổi")
plt.ylabel("Điểm số")
plt.show()
```
